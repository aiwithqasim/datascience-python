#### Introducing relplot()
- Create `"relational plots"`: scatter plots or line plot
    - Why use `relplot()` instead of `scatterplot()`?
- `relplot()` lets you create subplots in a single 

```python
sns.relplot(x="total_bill",y="tip",data=tips,
            kind="scatter")
plt.show()
```

### Subplots in columns

```python
sns.relplot(x="total_bill",y="tip",
            data=tips,
            kind="scatter",
            col="smoker")
plt.show()
```

#### Subplots in rows

```python
sns.relplot(x="total_bill",y="tip",
            data=tips,
            kind="scatter",
            row="smoker")
plt.show()
```

#### Subplots in rows and columns

```python
sns.relplot(x="total_bill",y="tip",
            data=tips,
            kind="scatter",
            col="smoker",
            row="time")
plt.show()
```

#### Wrapping columns

```python
sns.relplot(x="total_bill",y="tip",
            data=tips,
            kind="scatter",
            col="day",
            col_wrap=2)
plt.show()
```

#### Ordering columns

```python
sns.relplot(x="total_bill",y="tip",
            data=tips,kind="scatter",
            col="day",
            col_wrap=2,
            col_order=["Thur",
                       "Fri",
                       "Sat",
                       "Sun"])
plt.show()
```

### Customizing scatter plots
Till now we had covered scatter plots 

#### Scatter plot overview
Show relationship between two quantitative variables

***We've seen***:
- Subplots ( `col` and `row` )
- Subgroups with color ( `hue` )

***New Customizations***:
- Subgroups with point size and style
- Changing point transparency

***Use with both `scatterplot()` and `relplot()`***


#### Subgroups with point size

```python
sns.relplot(x="total_bill",y="tip",
            data=tips,
            kind="scatter",
            size="size")
plt.show()
```

#### Point size and hue

```python
sns.relplot(x= 'total_bill', y= 'tip',
          data = tips,
          kind = 'scatter',
          size = 'size',
          hue = 'size')
plt.show()
```

#### Subgroups with point style

```python
sns.relplot(x='total_bill', y='tip',
           data=tips,
           kind='scatter',
           hue='smoker',
           style='smoker')
plt.show()
```

#### Changing point transparency

```python
# set aplha between 0 and 1
sns.relplot(x='total_bill',y='tip',
           data=tips,
           kind='scatter',
           alpha=0.4)
plt.show()
```

#### Multiple observations per x-value

```python
#simple scatter plot
sns.relplot(x="hour", y="NO_2_mean",data=air_df_mean,kind="scatter")
plt.show()
```

#### Multiple observations per x-value

```python
# simple plot 
sns.relplot(x="hour", y="NO_2_mean",data=air_df_mean,kind="line")
plt.show()
```

***Shaded region is the confidence interval***
- Assumes dataset is a random sample
- 95% confident that the mean is within this interval
- Indicates uncertainty in our estimate

```python
# simple plot with confidence "Standard divation" 
sns.relplot(x="hour", y="NO_2_mean",data=air_df_mean,kind="line", ci='sd')
plt.show()
```

```python
# simple plot  with 'No' confidence
sns.relplot(x="hour", y="NO_2_mean",data=air_df_mean,kind="line", ci=None)
plt.show()
```

#### Subgroups by location

```python
# simple plot with No confidence & have Style & hue based on "location"
sns.relplot(x="hour", y="NO_2_mean",
            data=air_df_mean,
            kind="line",
            ci=None,
            style='location',
            hue='location')
plt.show()
```

#### Adding markers

```python
# old simple line plot with additional Markers feature 
sns.relplot(x="hour", y="NO_2_mean",
            data=air_df_mean,
            kind="line",
            ci=None,
            style='location',
            hue='location',
            markers=True)
plt.show()
```

#### Turning off line style

```python
# old simple line plot with additional Markers feature 
sns.relplot(x="hour", y="NO_2_mean",
            data=air_df_mean,
            kind="line",
            ci=None,
            style='location',
            hue='location',
            markers=True,
            dashes=False)
plt.show()
```

#### Creating two-factor subplots
- Let's continue looking at the student_data dataset of students in secondary school. Here, we want to answer the following question: does a student's first semester grade (`"G1"`) tend to correlate with their final grade (`"G3"`)?
- There are many aspects of a student's life that could result in a higher or lower final grade in the class. 
- For example, some students receive extra educational support from their school (`"schoolsup"`) or from their family (`"famsup"`), which could result in higher grades. 
- Let's try to control for ***these two factors by creating subplots based on whether the student received extra educational support from their school or family***.

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes","no"],
            row='famsup',
            row_order=['yes','no'])

# Show plot
plt.show()
```

## Miles Per Gallon ("M.P.G.") Dataset
we'll explore Seaborn's `mpg` dataset, which contains one row per car model and includes information such as the year the car was made, the number of miles per gallon ("M.P.G.") it achieves, the power of its engine (measured in "horsepower"), and its country of origin.

What is the relationship between the power of a car's engine (`"horsepower"`) and its fuel efficiency (`"mpg"`)? And how does this relationship vary by the number of cylinders (`"cylinders"`) the car has? Let's find out & explore the dataset.

```python
mpg.head()
```

```python
# Create scatter plot of horsepower vs. mpg
sns.relplot(x='horsepower',y='mpg',
            data=mpg,
            kind='scatter',
            size='cylinders')

# Show plot
plt.show()
```

```python
# Create scatter plot of horsepower vs. mpg
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders",
            hue='cylinders')

# Show plot
plt.show()
```

- Let's continue exploring Seaborn's mpg dataset by looking at the relationship between how fast a car can accelerate (`"acceleration"`) and its fuel efficiency (`"mpg"`). Do these properties vary by country of origin (`"origin"`)?

- Note that the `"acceleration"` variable is the time to accelerate from 0 to 60 ***miles per hour, in seconds***. Higher values indicate slower acceleration.

```python
# Create a scatter plot of acceleration vs. mpg
sns.relplot(x='acceleration',y='mpg',
            data=mpg,
            kind='scatter',
            hue='origin',
            style='origin')

# Show plot
plt.show()
```

we'll continue to explore Seaborn's `mpg` dataset, which contains one row per car model and includes information such as the year the car was made, its fuel efficiency (measured in "miles per gallon" or "M.P.G"), and its country of origin (USA, Europe, or Japan).

***QUESTION***:How has the average miles per gallon achieved by these cars changed over time? Let's use line plots to find out!

```python
# Create line plot
sns.relplot(x='model_year',y='mpg', data = mpg, kind='line')

# Show plot
plt.show()
```

***Question***

Which of the following is `NOT` a correct interpretation of this line plot?

***Possible Answers***

- The average miles per gallon has generally increased over time.
- The distribution of miles per gallon is smaller in 1973 compared to 1977.
- The 95% confidence interval for average miles per gallon in 1970 is approximately 16 - 19.5 miles per gallon.
- This plot assumes that our data is a random sample of all cars in the US, Europe, and Japan.

Answer: **B** Because The shaded region represents a confidence interval for the mean, not the distribution of the observations.

- we looked at how the average miles per gallon achieved by cars has changed over time. 
- Now let's use a line plot to visualize how the distribution of miles per gallon has changed over time.

```python
# Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci='sd')

# Show plot
plt.show()
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin",
            markers =True,
            dashes=False)

# Show plot
plt.show()
```
