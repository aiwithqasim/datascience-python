## While Loop
- ***While loop = repeated if statement***

- Numerically calculating model
- "repeating action until condition is met"
- Example
    - Error starts at 50
    - Divide error by 4 on every run
    - Continue until error no longer > 1

```python
error = 50.0
# 50.0 ---> True
# 12.5 ---> True
# 3.125 ---> True
# 0.78125 --> False
while error > 1:
    error = error / 4
    print(error)
```

- Continue until error no longer > 1
